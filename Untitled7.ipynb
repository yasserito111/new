```python
#we drop every column which does not give us many information
L=['MiscFeature','PoolQC','Fence','FireplaceQu','MasVnrType','Alley']
train.drop(L,axis=1,inplace=True)
test.drop(L,axis=1,inplace=True)
```

```python
numeric_data=train.select_dtypes(include=['int','float']).columns.tolist()
categorical_data=train.select_dtypes(include=['object']).columns.tolist()
```

```python
#we fill the missing parts that has influence
def filling(col):
    return col.fillna(col.mean())
train[numeric_data] = train[numeric_data].apply(filling)
numeric_data_test = [col for col in numeric_data if col != "SalePrice"]
test[numeric_data_test] = test[numeric_data_test].apply(filling)
```

```python
#now for the categorical data
for col in categorical_data:
    train[col].fillna(train[col].mode()[0], inplace=True)
    test[col].fillna(test[col].mode()[0], inplace=True)
```

```python
train[[col for col in train.columns if train[col].count() != 1460]].count()
```

Now we can start our visualizing


```python
plt.figure(figsize=(35, 35))
sns.heatmap(train[numeric_data].corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
useful_num_dt1=train.corr(numeric_only=True)['SalePrice'].sort_values(ascending=False).drop("SalePrice")
```

```python

useful_num_dt1 = useful_num_dt1[useful_num_dt1 >= 0]
useful_num_dt=useful_num_dt1.index.tolist()
```

```python
useful_num_dt
```

```python
categorical_data
train_encoded = pd.get_dummies(train[categorical_data], drop_first=True)
train_encoded['SalePrice'] = train['SalePrice']

# Compute correlations
correlations = train_encoded.corr()['SalePrice'].sort_values(ascending=False)
useful_cat_dt= correlations[correlations >= 0].index.tolist()
```

```python
train_final = pd.concat([train[useful_num_dt], train_encoded[useful_cat_dt]], axis=1)
test_encoded = pd.get_dummies(test[categorical_data], drop_first=True)
test_encoded = test_encoded.reindex(columns=train_encoded[useful_cat_dt].columns, fill_value=0)

test_final = pd.concat([test[useful_num_dt], test_encoded], axis=1)
```

```python
X_train = train_final
y_train = train['SalePrice']
```

Now we begin our fitting and predicting


```python
ss=StandardScaler()
lm=LinearRegression()
la=Lasso(alpha=0.1)
ridge=Ridge(alpha=0.3)
minmax=MinMaxScaler()
elsnet=ElasticNet()
```

```python

X_train_scaled = minmax.fit_transform(X_train)
test_final_scaled = minmax.transform(test_final)
y_train_log = np.log1p(y_train)
lm_ytest=lm.fit(X_train_scaled,y_train_log).predict(test_final_scaled)
lm_ytest=pd.DataFrame(lm_ytest)
lm_ytest.to_csv('lm_ytest.csv')
lm_ytest.head()
```
